```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fig,myaxes=plt.subplots(nrows=1,ncols=3)
x=np.linspace(10,300,10)
myaxes[0].plot(x,x*2)
myaxes[0].set_xlabel("X")
myaxes[0].set_ylabel("X*2")
myaxes[0].set_title("First")
myaxes[1].plot(x,x**2)
myaxes[1].set_xlabel("X")
myaxes[1].set_ylabel("X**2")
myaxes[1].set_title("second")
myaxes[2].plot(x,x**3)
myaxes[2].set_xlabel("X")
myaxes[2].set_ylabel("X**3")
myaxes[2].set_title("third")

plt.suptitle("Example of subbplots")
plt.tight_layout()
```

```python
fig,myaxes=plt.subplots(nrows=1,ncols=3)
p=2
title_list={2:"first",3:"Second",4:"Third"}
for a in myaxes:
    a.plot(x,x**p)
    a.set_xlabel("X")
    a.set_ylabel("X**"+str(p))
    a.set_title(title_list[p])
    p=p+1

fig.tight_layout()
```
