```python
def tip_Quality(total_bill,tip):
    if tip/total_bill>0.35:
        return "Very Good Tip"
    else:
        return "Bad Tip"
```

```python
df["tip_Quality"]=df.apply(lambda df:tip_Quality(df["total_bill"],df["tip"]),axis=1)
```
